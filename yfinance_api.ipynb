```python
import yfinance as yf

help(yf)
```

```python
tickerData = yf.Ticker("MSFT")
```

```python
###### Multiple tickers
tickers = yf.Tickers('msft aapl')
# ^ returns a named tuple of Ticker objects

# access each ticker using (example)
hist = tickers.history(period="1mo", interval="1d")
print(hist.columns)
display(hist["Close"])

display(hist.plot(figsize=(15,10)))
```
